```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('ready_for_tokenize')
data
```

```python
labelencoder = LabelEncoder()

data['Группа_обр'] = labelencoder.fit_transform(data['Классификация (Группа)'].tolist())
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['description_incident_ready'].tolist(), data['Группа_обр'].tolist(), test_size = 0.2, random_state=15 )
```

```python
tokenizer = Tokenizer(num_words=100000)
tokenizer.fit_on_texts(X_train)
```

```python
tokenizer.index_word
```

```python
vocab_size = len(tokenizer.index_word) + 1
print('Vocab Size:', vocab_size)
```

```python
X_train_token = tokenizer.texts_to_sequences(X_train)
X_test_token = tokenizer.texts_to_sequences(X_test)
```

```python
print("First Instance Text:\n")
print(X_train[0])
print("\nFirst Instance Total Words:", len(str(X_train[0]).split()))
```

```python
lens_sentences = [len(sentence.split()) for sentence in X_train]
max(lens_sentences)
```

```python
lens_sentences = [len(sentence.split()) for sentence in X_test]
max(lens_sentences)
```

```python
sequence_len = 30
X_train_token = pad_sequences(X_train_token, padding='post', maxlen=sequence_len)
X_test_token = pad_sequences(X_test_token, padding='post', maxlen=sequence_len)
print("First Instance Text Sequence:\n")
print(X_train_token[0])
print("\nFirst Instance Text Sequence Length:", len(X_train_token[0]))
```

```python
embedding_dim = 100
model = Sequential()
model.add(layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=sequence_len))
```

```python
model.add(layers.Conv1D(filters=128, kernel_size=5, activation='relu'))
model.add(layers.GlobalMaxPool1D())
```

```python
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(56, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
X_train_token = np.array(X_train_token,int)
y_train = np.array(y_train,int)
X_test_token = np.array(X_test_token,int)
y_test = np.array(y_test,int)

history = model.fit(X_train_token, y_train, epochs=20, validation_data=(X_test_token, y_test), batch_size=128)
```
